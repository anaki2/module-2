```python
# Name: Kiana Cooper
# File: casestudylab.py
# Description: This app accepts student names and GPAs and determines
#              if each student qualifies for the Dean's List (GPA >= 3.5)
#              or the Honor Roll (GPA >= 3.25). Enter 'ZZZ' as the last
#              name to quit the program.


def get_last_name():
    """Ask for and return the student's last name."""
    return input("Enter student's last name (or 'ZZZ' to quit): ")


def get_first_name():
    """Ask for and return the student's first name."""
    return input("Enter student's first name: ")


def get_gpa(first_name, last_name):
    """Ask for and return the student's GPA, re-asking if invalid."""
    while True:
        try:
            gpa = float(input(f"Enter {first_name} {last_name}'s GPA: "))
            return gpa
        except ValueError:
            # If the user types something that isn't a number, ask again
            print("Invalid GPA. Please enter a number (ex: 3.5).")


def check_qualification(first_name, last_name, gpa):
    """Check the student's GPA and print their qualification status."""
    if gpa >= 3.5:
        print(f"\n{first_name} {last_name} has made the Dean's List!\n")
    elif gpa >= 3.25:
        print(f"\n{first_name} {last_name} has made the Honor Roll!\n")
    else:
        print(f"\n{first_name} {last_name} did not qualify for the Dean's List or Honor Roll.\n")


def main():
    """Main function to keep processing students until ZZZ is entered."""
    print("=== Student GPA Qualifier ===\n")

    while True:
        last_name = get_last_name()

        if last_name == 'ZZZ':
            print("Exiting program. Goodbye!")
            break

        first_name = get_first_name()
        gpa = get_gpa(first_name, last_name)

        check_qualification(first_name, last_name, gpa)


# Run the program
main()
```
